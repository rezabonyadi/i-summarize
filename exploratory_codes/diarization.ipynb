```python
from resemblyzer import VoiceEncoder, preprocess_wav
from pathlib import Path
import numpy as np

# Read mp3 file, three people in a meeting.

fpath = './data/X2zqiX6yL3I.mp3'
wav = preprocess_wav(fpath)

# Embed the wav file using resemblyzer
encoder = VoiceEncoder('cpu')
_, cont_embeds, wav_splits = encoder.embed_utterance(wav, return_partials=True, rate=16)
```

```python
# Cluster embeded voices
from sklearn.cluster import SpectralClustering, KMeans
# clustering = SpectralClustering(n_clusters=3, assign_labels="discretize", random_state=0).fit(cont_embeds)
clustering = KMeans(n_clusters = 3).fit(cont_embeds)
```

```python
# Visualize clusters
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
# pca = PCA(n_components=20)
# pca_embd = tsne.fit_transform(test_features1)

# tsne = TSNE()
tsne = PCA()

X_trm = cont_embeds
# X_trm = pca_embd
tsne_embeds = tsne.fit_transform(X_trm)

print('TSNE Done')

x = tsne_embeds[:, 0]
y = tsne_embeds[:, 1]

plt.scatter(x, y, c=clustering.labels_)
plt.colorbar()
```

```python
# Calculate cluster membership scores as probability. Assume exponential distribution
cluster_prob = np.empty((cont_embeds.shape[0], clustering.cluster_centers_.shape[0]))
i = 0
j = 0
for i in range(cont_embeds.shape[0]):    
    embd = cont_embeds[i, :]
    for j in range(clustering.cluster_centers_.shape[0]):
        centre = clustering.cluster_centers_[j, :]
        cluster_prob[i, j] = np.exp(1.0/np.linalg.norm(embd-centre))
    cluster_prob[i,:]/=sum(cluster_prob[i, :])
```

```python
# Visualize who is talking in time

time_len = (wav.shape[0]/16000)

def moving_average(a, n=3):
    ret = np.cumsum(a, axis=0, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

ou = moving_average(cluster_prob, n=5) 
plt.figure(figsize=(20, 5))
plt.plot(np.linspace(0, time_len, ou.shape[0]), ou)
```

```python
# Creat labels in time

def create_labelling(labels,wav_splits):
    from resemblyzer import hparams
    sampling_rate = hparams.sampling_rate
    times = [((s.start + s.stop) / 2) / sampling_rate for s in wav_splits]
    labelling = []
    start_time = 0

    for i,time in enumerate(times):
        if i>0 and labels[i]!=labels[i-1]:
            temp = [str(labels[i-1]),start_time,time]
            labelling.append(tuple(temp))
            start_time = time
        if i==len(times)-1:
            temp = [str(labels[i]),start_time,time]
            labelling.append(tuple(temp))

    return labelling
  
labelling = create_labelling(clustering.labels_,wav_splits)
```
